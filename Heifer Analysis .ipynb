```python
import pandas as pd
import numpy as np
from scipy.stats import f_oneway
import plotly.graph_objects as go

# Load the dataset (replace with actual dataset and path)
file_path = "C:\\Users\\Samujjwal\\Desktop\\Animal Science data\\Longevity - Formatted-Souradip_analysis.xlsx"
df = pd.read_excel(file_path, sheet_name='MasterSheet')

# Convert the necessary columns to numeric, coercing errors to NaN
columns_to_analyze = ['AveLACT', 'Longevity', 'Time1stConception', 'ProductiveLife']
for col in columns_to_analyze:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Convert AMH column to numeric, coercing errors to NaN
df['AMH'] = pd.to_numeric(df['AMH'], errors='coerce')

# Filter out rows where AMH is NaN (we can't assign quartiles for NaN values)
df = df.dropna(subset=['AMH'])

# Divide into quartiles based on AMH
df['Quartile'] = pd.qcut(df['AMH'], 4, labels=['Q1', 'Q2', 'Q3', 'Q4'])

# Initialize dictionary to store results
results = {'Item': [], 'Q1 Mean (SEM)': [], 'Q2 Mean (SEM)': [], 'Q3 Mean (SEM)': [], 'Q4 Mean (SEM)': [], 'ANOVA (P-value)': []}

# Loop through each column and calculate the mean, SEM, and ANOVA p-value
for col in columns_to_analyze:
    q1 = df[df['Quartile'] == 'Q1'][col].dropna()
    q2 = df[df['Quartile'] == 'Q2'][col].dropna()
    q3 = df[df['Quartile'] == 'Q3'][col].dropna()
    q4 = df[df['Quartile'] == 'Q4'][col].dropna()

    # Check if the quartile columns have numeric values to avoid performing operations on empty data
    if len(q1) > 0 and len(q2) > 0 and len(q3) > 0 and len(q4) > 0:
        q1_mean_sem = f"{q1.mean():.2f}  {q1.sem():.2f}"
        q2_mean_sem = f"{q2.mean():.2f}  {q2.sem():.2f}"
        q3_mean_sem = f"{q3.mean():.2f}  {q3.sem():.2f}"
        q4_mean_sem = f"{q4.mean():.2f}  {q4.sem():.2f}"

        # Perform ANOVA to calculate the p-value
        anova_p_value = f_oneway(q1, q2, q3, q4).pvalue
    else:
        q1_mean_sem = q2_mean_sem = q3_mean_sem = q4_mean_sem = "N/A"
        anova_p_value = np.nan

    # Append results to the dictionary
    results['Item'].append(col)
    results['Q1 Mean (SEM)'].append(q1_mean_sem)
    results['Q2 Mean (SEM)'].append(q2_mean_sem)
    results['Q3 Mean (SEM)'].append(q3_mean_sem)
    results['Q4 Mean (SEM)'].append(q4_mean_sem)
    results['ANOVA (P-value)'].append(f"{anova_p_value:.3f}" if not np.isnan(anova_p_value) else "N/A")

# Create a DataFrame to store the results
results_df = pd.DataFrame(results)

# Create a Plotly table
fig = go.Figure(data=[go.Table(
    header=dict(values=list(results_df.columns),
                fill_color='paleturquoise',
                align='left'),
    cells=dict(values=[results_df[col] for col in results_df.columns],
               fill_color='lavender',
               align='left'))
])

# Show the plot
fig.show()
```
